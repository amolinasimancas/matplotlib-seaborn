```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randint(1,50,100)
```

```python
## Histograma
plt.hist(data,bins=10)
plt.show()
```

```python
plt.hist(data,bins=10,width=5,histtype='bar')
plt.show()
```

```python
plt.hist(data,bins=10,histtype='step')
plt.show()
```

```python
## Boxplot
plt.boxplot(data)
plt.show()
```

```python
data = np.append(data,200) ## simula un outlier
plt.boxplot(data,vert=False,patch_artist=True,notch=True)
plt.show()
```

```python
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=False)
plt.show()
```

```python
## Scatter
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30*np.random.rand(N))**2
colors = np.random.rand(N)
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='<',alpha=0.5)
plt.show()
```
