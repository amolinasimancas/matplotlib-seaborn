```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x=['A','B','C'],y=[1,2,3])
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(x=['A','B','C'],y=[1,2,3])
plt.show()
```

```python
sns.set(style='dark',palette='deep')
sns.barplot(x=['A','B','C'],y=[1,2,3])
plt.show()
```

```python
sns.set(style='dark',palette='dark',font_scale=1)
sns.barplot(x=['A','B','C'],y=[1,2,3])
plt.show()
```
